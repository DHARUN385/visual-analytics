```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
df = pd.read_csv("/content/heart.csv")

print("First 5 rows of dataset:")
print(df.head())

print("\nChecking for null values:")
print(df.isnull().sum())
df.rename(columns={'target': 'heart_disease'}, inplace=True)

plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='heart_disease', data=df)
sns.regplot(x='age', y='heart_disease', data=df, logistic=True, scatter=False, color='red')
plt.title('Age vs. Heart Disease')
plt.xlabel('Age')
plt.ylabel('Heart Disease (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='heart_disease', y='chol', data=df)
plt.title('Cholesterol Levels vs. Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
fig = px.histogram(df, x='exang', color='heart_disease', barmode='group',
labels={'exang': 'Exercise-Induced Angina', 'heart_disease': 'Heart Disease'},
title='Angina vs Heart Disease')
fig.update_layout(xaxis_title='Exercise-Induced Angina (1 = Yes, 0 = No)',
yaxis_title='Number of Patients')
fig.show()

print("\n🔚 Conclusion:")
print("- Age has a mild correlation with heart disease presence.")
print("- Cholesterol levels show some variation but are not the only factor.")
print("- Exercise-induced angina is strongly associated with heart disease.")
```
